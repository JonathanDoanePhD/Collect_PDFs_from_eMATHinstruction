```python
import pandas as pd
import numpy as np

import webbrowser as wb # to open webpage
import time # to pause Python for given time
import msvcrt # to pause Python until key is pressed
import wget # to download file from URL
import pyautogui as pag
import pynput
from pynput.mouse import Button
from pynput.mouse import Controller as ms_Controller
from pynput.keyboard import Key
from pynput.keyboard import Controller as kb_Controller
ms = ms_Controller()
kb = kb_Controller()

safe_position  = (20,10)
input_position = (620, 270) #(400, 180)
# input_scroll   = (0,-1000)

wait = 1
tap = 0.1


# Common commands

def Ctrl(*in_tasks, in_wait=wait*tap):
    pag.hotkey('ctrl', *in_tasks)
    time.sleep(in_wait)

def Alt(*in_tasks, in_wait=wait*tap):
    pag.hotkey('alt', *in_tasks)
    time.sleep(in_wait)

def Shift(*in_tasks, in_wait=wait*tap):
    pag.hotkey('shift', *in_tasks)
    time.sleep(in_wait)

def Repeat(in_int, in_task, in_wait=wait*tap):
    for i in range(in_int):
        eval(str(in_task))
        time.sleep(in_wait)

def Enter(in_int=1, in_wait=wait*tap):
    Repeat(in_int, 'kb.tap(Key.enter)')

def Tab(in_int=1, in_wait=wait*tap):
    for i in range(in_int):
        kb.tap(Key.tab)
        time.sleep(in_wait)

def Delete(in_int=1, in_wait=wait*tap):
    for i in range(in_int):
        kb.tap(Key.delete)
        time.sleep(in_wait)

def Backspace(in_int=1, in_wait=wait*tap):
    for i in range(in_int):
        kb.tap(Key.backspace)
        time.sleep(in_wait)

def Select_All(in_wait=wait*tap):
    Ctrl('a')
    time.sleep(in_wait)

def Copy(in_wait=wait*tap):
    Ctrl('c')
    time.sleep(in_wait)

def Paste(in_wait=wait*tap):
    Ctrl('v')
    time.sleep(in_wait)

def Esc(in_wait=wait*tap):
    pag.hotkey('esc')
    time.sleep(in_wait)

def F3(in_search, in_enter=0, in_wait=wait*tap):
    kb.tap(Key.f3)
    time.sleep(in_wait)
    Backspace()
    kb.type(in_search)
    time.sleep(in_wait)
    for i in range(in_enter):
        Enter()
    pag.hotkey('esc')
    time.sleep(in_wait)

def Scroll_Top(in_wait=wait*tap):
    pag.hotkey('home')
    time.sleep(in_wait)

def Scroll_Bottom(in_wait=wait*tap):
    pag.hotkey('end')
    time.sleep(in_wait)


# Browser commands

def Go_Back(in_int=1):
    for i in range(in_int):
        Alt('left')
        time.sleep(wait_tab)
    time.sleep(wait_website)

def Close_Tab():
    Ctrl('w')


# JL commands

def Save():
    Ctrl('s')
    time.sleep(wait_file)

def New_Cell(): # start from selected cell(s)
    kb.type('b')
    time.sleep(wait_tab)

def Delete_Cell(): # start from selected cell(s)
    kb.type('dd')
    time.sleep(wait_tab)

def Run_Cell(in_int=1):
    Ctrl('enter')
    time.sleep(wait_cell)

def Right_JL_Tab(in_int=1):
    for i in range(in_int):
        Ctrl('shift', ']')
        time.sleep(wait_tab)
    time.sleep(wait_cell)

def Left_JL_Tab(in_int=1):
    for i in range(in_int):
        Ctrl('shift', '[')
        time.sleep(wait_tab)
    time.sleep(wait_cell)

def Undo_Autocomplete():
    Ctrl('delete')

def Copy_Output(in_var):
    pd.DataFrame([in_var]).to_clipboard(index=False,header=False)

def Await_Answer():
    msvcrt.getch()
    time.sleep(wait_cell)


# Navigation commands

def To_JL(): # start from browser
    Ctrl('tab')

def To_Browser(): # start from JL
    Ctrl('shift', 'tab')

def Directory_Bar():
    Ctrl('l')


def Execute_Clipboard():
    # Left_JL_Tab()
    # ms.position = input_position
    # time.sleep(wait_mouse)
    # # ms.scroll(*input_scroll)
    # Scroll_Bottom()
    # time.sleep(wait_mouse)
    # ms.click(Button.left)
    # time.sleep(wait_mouse)
    # time.sleep(wait_execute)
    # Enter()
    time.sleep(wait_execute)
    Paste()
    Enter()

def Get_Variable(in_var, in_search): # start from browser, input strings
    Select_All()
    Copy()
    To_JL()
    time.sleep(wait_cell)
    Right_JL_Tab()
    New_Cell()
    Enter()
    kb.type(in_var)
    kb.type(' = """')
    Undo_Autocomplete()
    Paste()
    kb.type('"""')
    Undo_Autocomplete()
    kb.type('.count("')
    kb.type(in_search)
    kb.type('")')
    Enter()
    kb.type('Copy_Output(')
    kb.type(in_var)
    kb.type(')')
    Run_Cell()
    Enter()
    Select_All()
    Delete()
    kb.type('Execute_Clipboard()')
    Run_Cell()
    Left_JL_Tab()


# Start and finish

def Safe_Start():
    time.sleep(2)
    kb.release(Key.ctrl)
    kb.release(Key.enter)
    ms.position = safe_position
    New_Cell()
    Right_JL_Tab()
    Left_JL_Tab()
    Delete_Cell()

def Start():
    Safe_Start()
    # Copy imports and functions
    kb.tap(Key.up)
    Enter()
    Select_All()
    Copy()
    # Load imports and functions in execution JL-tab
    Right_JL_Tab()
    Enter()
    Paste()
    Run_Cell()
    # Delete_Cell()
    Left_JL_Tab()
    kb.tap(Key.down)

def Finish():
    To_JL()
    Right_JL_Tab()
    Select_All()
    Delete_Cell()
    Left_JL_Tab()
```
