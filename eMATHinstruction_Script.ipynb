```python
import pandas as pd
import webbrowser as wb # to open webpage
import time # to pause Python for given time
import msvcrt # to pause Python until key is pressed
import wget # to download file from URL
import pyautogui as pag
import pynput
from pynput.mouse import Button
from pynput.mouse import Controller as ms_Controller
from pynput.keyboard import Key
from pynput.keyboard import Controller as kb_Controller
ms = ms_Controller()
kb = kb_Controller()

safe_position  = (20,10)
input_position = (620, 270) #(400, 180)
# input_scroll   = (0,-1000)

website = "https://www.emathinstruction.com/courses/n-gen-math-7/" #"https://www.emathinstruction.com/"
curriculum = 'N-GEN_MATH'
courses = ['7', '8', 'ALGEBRA_1']

wait_cell    = 0.5
wait_copy    = 0.2
wait_execute = 2.0 # 2
wait_f3      = 0.5
wait_file    = 2.0 # 2
wait_mouse   = 0.2
wait_popup   = 0.5
wait_scroll  = 1.0
wait_tab     = 0.05
wait_website = 3.0 # 3

# wait_cell    = 2
# wait_copy    = 2
# wait_execute = 2
# wait_f3      = 2
# wait_file    = 2
# wait_mouse   = 2
# wait_popup   = 2
# wait_tab     = 2
# wait_website = 2

# wait = [wait_cell, wait_copy, wait_execute, wait_f3, wait_file, wait_mouse, wait_popup, wait_tab, wait_website]

# def Slow(in_truth=True, in_wait=2):
#     if bool(in_truth) == True:
#         for i in range(len(wait)):
#             wait[i] = in_wait
#     else:
#         pass
#     return None


# Common commands

def Enter(in_int=1):
    for i in range(in_int):
        kb.tap(Key.enter)
        time.sleep(wait_tab)
    return None

def Tab(in_int=1):
    for i in range(in_int):
        kb.tap(Key.tab)
        time.sleep(wait_tab)
    return None

def Delete(in_int=1):
    for i in range(in_int):
        kb.tap(Key.delete)
        time.sleep(wait_tab)
    return None

def Backspace(in_int=1):
    for i in range(in_int):
        kb.tap(Key.backspace)
        time.sleep(wait_tab)
    return None


def Ctrl(*in_tasks):
    pag.hotkey('ctrl', *in_tasks)
    return None

def Alt(*in_tasks):
    pag.hotkey('alt', *in_tasks)
    return None

def Shift(*in_tasks):
    pag.hotkey('shift', *in_tasks)
    return None



def Select_All():
    Ctrl('a')
    time.sleep(wait_copy)
    return None

def Copy():
    Ctrl('c')
    time.sleep(wait_copy)
    return None

def Paste():
    Ctrl('v')
    time.sleep(wait_copy)
    return None


def F3(in_search, in_enter=0):
    kb.tap(Key.f3)
    time.sleep(wait_f3)
    Backspace()
    # time.sleep(wait_f3)
    kb.type(in_search)
    time.sleep(wait_f3)
    for i in range(in_enter):
        Enter()
    pag.hotkey('esc')
    time.sleep(wait_f3)
    return None

def Scroll_Top():
    pag.hotkey('home')
    time.sleep(wait_scroll)
    return None

def Scroll_Bottom():
    pag.hotkey('end')
    time.sleep(wait_scroll)
    return None


# Browser commands

def Go_Back(in_int=1):
    for i in range(in_int):
        Alt('left')
        time.sleep(wait_tab)
    time.sleep(wait_website)
    return None

def Close_Tab():
    Ctrl('w')
    return None


# JL commands

def Save():
    Ctrl('s')
    time.sleep(wait_file)
    return None

def New_Cell(): # start from selected cell(s)
    kb.type('b')
    time.sleep(wait_tab)
    return None

def Delete_Cell(): # start from selected cell(s)
    kb.type('dd')
    time.sleep(wait_tab)
    return None

def Run_Cell(in_int=1):
    Ctrl('enter')
    time.sleep(wait_cell)
    return None

def Right_JL_Tab(in_int=1):
    for i in range(in_int):
        Ctrl('shift', ']')
        time.sleep(wait_tab)
    time.sleep(wait_cell)
    return None

def Left_JL_Tab(in_int=1):
    for i in range(in_int):
        Ctrl('shift', '[')
        time.sleep(wait_tab)
    time.sleep(wait_cell)
    return None

def Undo_Autocomplete():
    Ctrl('delete')
    return None

def Copy_Output(in_var):
    pd.DataFrame([in_var]).to_clipboard(index=False,header=False)
    return None

def Await_Answer():
    msvcrt.getch()
    time.sleep(wait_cell)
    return None


# Navigation commands

def To_JL(): # start from browser
    Ctrl('tab')
    time.sleep(wait_tab)
    return None

def To_Browser(): # start from JL
    Ctrl('shift', 'tab')
    time.sleep(wait_tab)
    return None

def Directory_Bar():
    Ctrl('l')
    time.sleep(wait_tab)
    return None

# Main tasks

def To_Course(in_course, in_curriculum = curriculum): #Start from website
    if in_curriculum.upper() == 'N-GEN_MATH':
        t_curriculum = 'N-Gen Mathâ„¢'
    else:
        t_curriculum = in_curriculum
    t_course = in_course.replace('_',' ')
    # Open FOR TEACHERS popup
    F3('FOR TEACHERS')
    # Navigate to Course
    F3(t_curriculum +' '+ t_course)
    Enter()
    time.sleep(wait_website)
    return None

def To_Unit(in_unit, in_search=' - '):
    F3(in_search)
    # time.sleep(wait_popup)
    Tab(int(in_unit)) # unit
    time.sleep(wait_popup)
    Enter()
    time.sleep(wait_website)
    return None

def To_Lesson(in_lesson, in_bad=0, in_search='Lesson'):
    F3(in_search)
    Tab(2+int(in_lesson*(8-in_bad))) # lesson
    Enter()
    time.sleep(wait_popup)
    Tab() # accept conditions
    Enter()
    time.sleep(wait_website)
    return None

def To_Answer(in_lesson, in_bad=0, in_search='Lesson'):
    F3(in_search)
    Tab(5+int(in_lesson*(8-in_bad))) # lesson
    Enter()
    time.sleep(wait_popup)
    Tab() # accept conditions
    Enter()
    time.sleep(wait_website)
    return None

    # # Download the open tab then close it
    # Download(in_curriculum, in_course, in_unit, in_lesson)
    # Close_Tab()


def Execute_Clipboard():
    # Left_JL_Tab()
    # ms.position = input_position
    # time.sleep(wait_mouse)
    # # ms.scroll(*input_scroll)
    # Scroll_Bottom()
    # time.sleep(wait_mouse)
    # ms.click(Button.left)
    # time.sleep(wait_mouse)
    # time.sleep(wait_execute)
    # Enter()
    time.sleep(wait_execute)
    Paste()
    Enter()
    return None

def Get_Variable(in_var, in_search): # start from browser, input strings
    Select_All()
    Copy()
    To_JL()
    time.sleep(wait_cell)
    Right_JL_Tab()
    New_Cell()
    Enter()
    kb.type(in_var)
    kb.type(' = """')
    Undo_Autocomplete()
    Paste()
    kb.type('"""')
    Undo_Autocomplete()
    kb.type('.count("')
    kb.type(in_search)
    kb.type('")')
    Enter()
    kb.type('Copy_Output(')
    kb.type(in_var)
    kb.type(')')
    Run_Cell()
    Enter()
    Select_All()
    Delete()
    kb.type('Execute_Clipboard()')
    Run_Cell()
    Left_JL_Tab()
    # Enter()
    # Await_Answer()
    # Delete_Cell()
    return None

def Get_Units():
    Get_Variable('num_units', ' - ')
    return None

def Get_Lessons():
    Get_Variable('num_lessons', 'Lesson')
    return None


def Download(in_course, in_unit, in_lesson, in_answer=False, in_curriculum=curriculum): # Start from new download tab
    if in_answer == False:
        t_answer = 'Problems\\'
    else:
        t_answer = 'Solutions\\'
    Save()
    Directory_Bar()
    kb.type('C:\\Users\\kitsu\\Documents\\0TLCSD\\eMATHinstruction\\' +t_answer+ in_curriculum +'_'+ in_course)
    time.sleep(wait_copy)
    Enter()
    time.sleep(wait_copy)
    kb.tap(Key.f6)
    time.sleep(wait_tab)
    Ctrl('enter')
    time.sleep(wait_file)
    # Tab(8) # download button
    # Enter()
    # time.sleep(wait_file)
    # Tab(6) # directory
    # Enter()
    # kb.type('C:\\Users\\kitsu\Documents\\0TLCSD\\eMATHinstruction\\' +in_curriculum+'_'+in_course)
    # Enter()
    # time.sleep(wait_file)
    # Tab(6) # file name
    # kb.type(str(in_curriculum) +'_'+ str(in_course) +'_Unit'+ str(in_unit) +'_Lesson'+ str(in_lesson))
    # Enter()
    # time.sleep(wait_file)
    # if in_answer = False:
    #     t_answer = 'Problems\\'
    # else:
    #     t_answer = 'Solutions\\'
    # output_directory = 'C:\\Users\\kitsu\\Documents\\0TLCSD\\' + t_answer
    # kb.tap(Key.f6)
    # time.sleep(wait_f3)
    # Copy()
    # To_JL()
    # Right_JL_Tab()
    # New_Cell()
    # Enter()
    # kb.type('Execute_Clipboard()')
    # Run_Cell()
    # Left_JL_Tab()
    # # output_directory = 'P_' +str(in_curriculum) +'_'+ str(in_course) +'_Unit'+ str(in_unit) +'_Lesson'+ str(in_lesson)
    # wget.download(url, out=output_directory)
    return None


# Start and finish

def Safe_Start():
    time.sleep(2)
    kb.release(Key.ctrl)
    kb.release(Key.enter)
    ms.position = safe_position
    New_Cell()
    Right_JL_Tab()
    Left_JL_Tab()
    Delete_Cell()
    return None

def Start():
    Safe_Start()
    # Copy imports and functions
    kb.tap(Key.up)
    Enter()
    Select_All()
    Copy()
    # Load imports and functions in execution JL-tab
    Right_JL_Tab()
    Enter()
    Paste()
    Run_Cell()
    # Delete_Cell()
    Left_JL_Tab()
    kb.tap(Key.down)
    return None

def Finish():
    To_JL()
    Right_JL_Tab()
    Select_All()
    Delete_Cell()
    # Save()
    Left_JL_Tab()
    # Save()
    return None
```
